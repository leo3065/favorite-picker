```python
import os
```

```python
imgs = os.listdir('image/offical_thumbnail')
```

```python
items = []
for img in imgs:
    name = img.removesuffix('.png')
    items.append(
        '{'
        f"id: '{name.lower().replace("'", "-")}', "
        f"name: '{name.lower().replace("'", "-").replace("'", "\\'")}', "
        f"image: 'image/offical_thumbnail/{img.replace("'", "\\'")}'"
        '}'
    )
print(',\n'.join(items))
```

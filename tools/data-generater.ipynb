```python
import os
```

```python
imgs = os.listdir('../image/official_thumbnail')
```

```python
items = []
for img in imgs:
    name = img.removesuffix('.png')
    items.append(
        '{'
        f"id: '{name.lower().replace("'", "-")}', "
        f"name: '{name.replace("_", " ").replace("'", "\\'")}',\n\t"
        f"image: 'image/official_thumbnail/{img.replace("'", "\\'")}'"
        '}'
    )
print(',\n'.join(items))
```
